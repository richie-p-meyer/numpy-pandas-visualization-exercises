```python
from pydataset import data
import pandas as pd
import numpy as np
```

```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

df
```

```python
df['name'].str.startswith('S')
```

```python
df.loc[df['name'].str.startswith('S'),['name','math']]
```

```python
df['passing_english'] = df.english>=70
```

```python
df.sort_values(by = 'passing_english')
#sorted by index after the col passed in
```

```python
df.sort_values(by = ['passing_english','name'])
#Sorted by name second, third index
```

```python
df.sort_values(by = ['passing_english','english'])
```

```python
df['overall_grade']= round((df['math']+df['english']+df['reading'])/3,0)
```

```python
mpg = data('mpg')
```

```python
mpg.shape
#234 rows, 11 cols
```

```python
mpg.info()
```

```python
mpg.describe()
```

```python
#Rename the cty column to city
mpg = mpg.rename(columns = {'cty':'city'})
```

```python
mpg = mpg.rename(columns = {'hwy':'highway'})
```

```python
mpg[mpg.city>mpg.highway]
#No cars get better mpg during city driving than highway
```

```python
mpg['mileage_difference'] = mpg['highway']-mpg['city']
mpg.head()
```

```python
mpg.sort_values(by='mileage_difference',ascending=False).head()
#Highest mileage_difference
```

```python
mpg[mpg['class']=='compact'].sort_values(by = 'highway').head()
#Worst highway mileage
```

```python
mpg[mpg['class']=='compact'].sort_values(by = 'highway',ascending=False).head()
#Best highway mileage
```

```python
#Add column that calculates city and highway mpg average
mpg['average_mileage'] = (mpg.city +mpg.highway)/2
mpg.head()
```

```python
#Which dodge car has the worst average mileage? 
mpg[mpg['manufacturer']=='dodge'].sort_values(by='average_mileage').head()
```

```python
#Which dodge car has the best average mileage? 
mpg[mpg['manufacturer']=='dodge'].sort_values(by='average_mileage', ascending = False).head()
```

```python
#How many animals are hoppers that are above the median speed? What percentage is this?
median_speed = mammals['speed'].median()
mammals[(mammals['hoppers']==True) & (mammals['speed'] > median_speed)][['hoppers']].count()/mammals.shape[0]*100
```
