```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from random import randint
import math
%config InlineBackend.figure_format = 'retina'
```

```python
# Plot y = x2 - x + 2
x = range(-100,100)
y = [xx**2-xx+2 for xx in x]
plt.xlabel('x')
plt.ylabel('y')
plt.title('$x^2 -x +2')
plt.plot(x,y,c='green')
plt.annotate('Origin', xy = (0,0), xytext = (0,5),arrowprops = {'facecolor':'blue'})
plt.show()
```

```python
plt.plot([xx**.5 for xx in x if xx > 0])
plt.title('$y=sqr(x)$')
```

```python
plt.plot(x, [xx**3 for xx in x])
plt.title('$y = x^3$')
```

```python
y = [2**xx for xx in x]
plt.plot(x,y)
plt.title('$y=2^x$')
plt.xlim(0,5)
plt.ylim(0,50)
```

```python
plt.plot(x, [1/(xx+1) for xx in x])
plt.title('$y=1/(x+1)$')
```

```python
x = range(0,100)

plt.subplot(221)
plt.plot(x, [xx**.5 for xx in x])
plt.title('$y=sqr(x)$')


plt.subplot(222)
plt.plot(x, [xx**3 for xx in x])
plt.title('$x^3$')


plt.subplot(223)
plt.plot(x, [2**xx for xx in x])
plt.title('$y=2^x$')


plt.subplot(224)
plt.plot(x, [1/(xx+1) for xx in x])
plt.title('$y=1/(x+1)$')


plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
```

```python
plt.figure(figsize = (8,4))
plt.plot([xx**.5 for xx in x], c = 'green', label = '$y=\sqrt{x}$')
plt.plot([xx**3 for xx in x], c = 'blue', label = '$x^3$')
plt.plot([2**xx for xx in x], c = 'yellow', label = '$y=2^x$')
plt.plot([1/(xx+1) for xx in x if xx != -1], c = 'orange', label ='$y=1/(x+1)$')
plt.title('Four different math equations')
plt.ylim(0,10)
plt.xlim(0,10)
plt.legend()
```
