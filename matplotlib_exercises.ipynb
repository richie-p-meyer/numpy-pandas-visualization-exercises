```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from random import randint
import math
```

```python
# Plot y = x2 - x + 2
x = range(-50,50)
y = [xx**2-xx+2 for xx in x]
plt.plot(x,y,c='green')
plt.annotate('Origin', xy = (0,0), xytext = (0,500),arrowprops = {'facecolor':'blue'})
plt.show()
```

```python
x = range(0,100)

plt.subplot(221)
plt.plot(x, [xx**.5 for xx in x])
plt.title('$y=sqr(x)$')

plt.subplot(222)
plt.plot(x, [xx**3 for xx in x])
plt.title('$x^3$')

plt.subplot(223)
plt.plot(x, [2**xx for xx in x])
plt.title('$y=2^x$')

plt.subplot(224)
plt.plot(x, [1/(xx+1) for xx in x])
plt.title('$y=sqr(x)$')
```
