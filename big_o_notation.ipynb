```python
import math
import matplotlib.pyplot as plt
```

```python
x = range(1,10)

plt.plot(x,[0*r +1 for r in x], label = '$O(1)$')

plt.plot(x,[math.log(r) for r in x], label = '$O(logn)$')



plt.plot(x,[r for r in x], label = '$O(n)$')


plt.plot(x,[r * math.log(r) for r in x], label = '$O(nlog(n))$')


plt.plot(x,[r**2 for r in x], label = '$O(n^2)$')


plt.plot(x,[2**r for r in x], label = '$O(2^n)$')


plt.plot(x,[math.factorial(r) for r in x], label = '$O(n!)$')


plt.plot(x,[r**r for r in x], label = '$O(n^n)$')
plt.legend()
plt.show()
```
